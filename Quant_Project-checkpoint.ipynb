```python
#Calculating the portfolio returns and volatility.


stocks = ['UBS', 'UAL', 'UBER']
weights = [0.5, 0.25, 0.25]
total_investment = 10000
investment_amounts = [weight * total_investment for weight in weights]
data = pd.read_csv('Portfolio.csv')
data = data.set_index(pd.DatetimeIndex(data['Date'].values))
data.drop(columns=['Date'], axis=1, inplace=True)
portfolio_returns = (data.pct_change()*investment_amounts).sum(axis=1)
portfolio_volatility = portfolio_returns.std()*np.sqrt(252)
print("Portfolio returns: ", portfolio_returns)
print("Portfolio volatility: ", portfolio_volatility)
```

```python
#Investing 50% of the funds to UBS, and 25% each to United Airline(uAL) and UBER.


assets = ['UBS', 'UBER', 'UAL']
weights = np.array([0.5, 0.25, 0.25])
```

```python
#Setting the start date and the end date for fetching the data.


stockStartDate = '2022-08-19'
stockEndDate = '2024-08-19'
```

```python
#Accessing only the closing prices for the stocks in our portfolio.


df = pd.DataFrame()
for stock in assets:
    df[stock] = yf.download(stock, start=stockStartDate, end=stockEndDate)['Adj Close']
df.head()
```

```python
#Potting the adjusted closing prices for the stocks in the portfolio.


title = 'Portfolio Adj. Close Price History'
my_stocks = df
for c in my_stocks.columns.values:
    plt.plot(my_stocks[c], label=c)
plt.title(title)
plt.xlabel('Date', fontsize=18)
plt.ylabel('Adj. Price USD ($)', fontsize=18)
plt.legend(my_stocks.columns.values, loc='upper left')
plt.show()
```

```python
#Getting the relative price change of the stocks.


returns = df.pct_change()
returns.head()
```

```python
#Calculating the annual covariance of the stocks.


cov_matrix_annual = returns.cov()*252
cov_matrix_annual
```

```python
#Here we are calculating the variance of the portfolio.


port_variance = np.dot(weights.T, np.dot(cov_matrix_annual, weights))
port_variance
```

```python
#We are calculating the standard deviation/volatility of te portfolio.


port_volatility = np.sqrt(port_variance)
port_volatility
```

```python
# Here we are calculating the simple annual return of our portfolio.


portfolioSimpleAnnualReturn = np.sum(returns.mean()*weights)*252
portfolioSimpleAnnualReturn
```

```python
#This is a summary of the various performance indicators for the portfolio, containing information about
# their expected annual return, annual volatility and annual variance.


percent_var = str(round(port_variance, 2)*100)+ '%'
percent_vol = str(round(port_volatility, 2)*100)+'%'
percent_ret = str(round(portfolioSimpleAnnualReturn, 2)*100)+'%'
print('Expected annual return: '+ percent_ret)
print('Annual volatility/risk: ' + percent_vol)
print('Annual variance: ' + percent_var)
```

```python
#Getting the correlational matrix of the portfolio.
#We can see that all the stocks in our portfolio have low correlational values with each other.

returns.corr()
```

Correlational values are important for the analysis of a portfolio because they tell us how the stocks would be affected if one of the stocks gains or loses its value. If the correlational  values are high, it means that if one stock's prices increases or decreases for a certain percent, other stocks would also experience a similar amount of increase/decrease in their value. It usually happens in portfolios having stocks from the same industry. In our case, the correlational values are low which means that an increased volatility in the price of one stock wouldn't affect the other stocks that much as the portfolio is diversified.
